```python
import pandas as pd
import pylab
import matplotlib.pyplot as plt

%pylab inline
```

```python
import glob
csvs = glob.glob("data/*.csv")

dfs = [pd.read_csv(csv) for csv in csvs]
for df in dfs:
    df['@MeasurementDateGMT'] = pd.to_datetime(df['@MeasurementDateGMT'], format="%Y-%m-%d %H:%M:%S")
dfs = [df.set_index("@MeasurementDateGMT") for df in dfs]
```

```python
filtered_dfs = [df for df in dfs if 'NO2' in df['@SpeciesCode'].unique()]
len(filtered_dfs)
big_df = pd.concat(filtered_dfs)
big_df_filtered = big_df[big_df['@SpeciesCode'] == "NO2"]
```
